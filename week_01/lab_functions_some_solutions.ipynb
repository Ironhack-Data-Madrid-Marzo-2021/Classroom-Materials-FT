```python
def pangram(string):
    alphabet = list(map(chr, range(97, 123)))
    
    for letter in alphabet:
        if letter not in string.lower():
            return False
    
    return True
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.
`Valid special characters: # @ ! $ % & ( ) ^ * [ ] { }`

```python
def check_pass(password):
    lower_case = False
    upper_case = False
    number = False
    special = False
    special_characters = '#@!$%&()^*[]{}.'
    if len(password) < 8:
        return False
    for char in password:
        print(f"checking character {char}")
        if char.islower():
            lower_case = True
        elif char.isupper():
            upper_case = True
        elif char.isdigit():
            number = True
        elif char in special_characters:
            special =True
        
        if lower_case and upper_case and number and special:
            return True
    
    
    return lower_case and upper_case and number and special
    # return False
```

```python
# This will test your function 
test_pass(check_pass)
```

## Prime numbers

Prime numbers cannot be divided by *anyone*

12 not prime: $12 = 3*4$  
25 not prime  $25 = 5*5$  
11 prime  
13 prime  

```python
import math
```

```python
def is_prime(n):
    # we only need to try up to ~ square root of n
    biggest_try = math.ceil(n ** 0.5)
    
    for i in range(2, biggest_try + 1):
        # if n can be divided by i, n is NOT prime
        if n % i == 0:
            # print(f"{n} can be divided by {i}")
            return False
    
    # if we got here, no divisor of n was found
    return True
```
