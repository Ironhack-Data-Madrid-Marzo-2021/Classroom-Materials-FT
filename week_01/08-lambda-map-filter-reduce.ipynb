##Funciones lambda

Una funci贸n lambda es una funci贸n an贸nima, sin identificador, que puede declararse in situ. Dado que podemos usar funciones como argumentos y asignarlas a variables, la notaci贸n que posibilita la creaci贸n de funciones lambda es muy pr谩ctica en muchos casos.

### Sintaxis de una lambda
```python

```

### Ejemplos de lambda

````python
lambda x: x **2 #Devuelve el cuadrado de un n煤mero
lambda x, y: x + y #devuelve la suma de los dos argumentos
lambda x: x[0] #devuelve es primer elemento del argumento
lambda x,y : x if x > y else y #devuelve el mayor de los argumentos 
````

Tambi茅n es posible usar estas funciones an贸nimas como argumentos en la llamada a otras funcoines. Por ejemplo, en una llamada a sorted() o list.sort() podemos usar una funci贸n lambda como argumento para el par谩metro key:

````python
distancias = [("Madrid", 331), ("C贸rdoba", 117), ("Barcelona",798)]
````

````python
#Ordenamos por la distancia, de menor a mayor by default 
sorted(distancias, key= lambda x: x[1])
````

````python
#Ordenamos por distancia pero de mayor a menor, con reverse = True
sorted(distancias, key = lambda x: x[1], reverse= True)
````

````python
#Ordenamos alfab茅ticamente utilizando el primer elemento x[0]
sorted(distancias, key= lambda x: x[0])
````

````python
#Ordenamos alfab茅ticamente de final a principio utilizando reverse = True
sorted(distancias, key= lambda x: x[0], reverse = True)
````

Por 煤ltimo, si queremos devolver como resultado de una funci贸n a otra funci贸n podemos usar cualquier funci贸n conocida en el 谩mbito de la funci贸n que realiza el retorno, o usar una funci贸n lambda:

````python
def comparaciones (tipo):
    
    def mayor (a,b):
        if a > b:
            return a
        else:
            return b
        
    if tipo == "Mayor":
        return mayor
    
    if tipo == "Menor":
        return lambda x,y: x if x < y else y
````

````python
def comparaciones_2 (tipo):
    """
    Esta funci贸n devuelve otra funci贸n.
    Devuelve la funci贸n mayor si como argumento entra "Mayor" en forma de string
    Devuelve menor cuando en el argumento entra "Menor" en forma de string
    """
    def mayor (a,b):
        if a > b:
            return a
        else:
            return b
    
    def menor (a,b):
        if a < b:
            return a
        else:
            return b
        
    if tipo == "Mayor":
        return mayor
    
    if tipo == "Menor":
        return menor
````

````python
def comparaciones_3 (tipo, a, b):
    if tipo == "Mayor":
        if a > b:
            return a
        else:
            return b
        
    else:
        if a < b:
            return a
        else:
            return b
````

````python
el_mayor = comparaciones("Mayor")(19,20)
el_menor = comparaciones("Menor")
````

###Reduce

Esta funci贸n no es una funci贸n predefinida, lo cual implica importar antes la biblioteca functools para su uso. La funci贸n reduce() aplica de forma acumulativa la funci贸n f() sobre los elementos del iterable. El acumulador puede inicializarse con un tercer argumento opcional. El resultado es el valor final del acumulador.
```python
reduce(funci贸n, iterable[, inicial])
```

````python
from functools import reduce
````

锔 Reduce es un poco m谩s complicado de entender que map() y filter() as铆 que veamos paso a paso el siguiente ejemplo 锔

````python
def suma(x,y):
    return x +y
lista = [2,4,7,3]
print(reduce(suma,lista))
````

#### 驴Qu茅 ha pasado?
- Empezamos con una lista [2, 4, 7, 3] y pasamos la funci贸n suma(x, y) a reduce() junto con esta lista, sin un valor inicial

- reduce() llama a suma(2, 4), y suma() devuelve 6

- reduce() llama a suma(6, 7) (resultado de la llamada anterior a suma() y el siguiente elemento de la lista como par谩metros), y suma() devuelve 13

- reduce() llama a suma(13, 3), y suma() devuelve 16

Como no quedan m谩s elementos en la secuencia, reduce() devuelve 16

Factorial:     
Cantidad que resulta de la multiplicaci贸n de determinado n煤mero natural por todos los n煤meros naturales que le anteceden excluyendo el cero; se representa por n!

````python
def factorial(n):
    """Funci贸n que calcula el factorial de un n煤mero entero positivo.
    Par谩metros
    n: Es un entero positivo.
    Devuelve el factorial de n.
    """
    tmp = 1
    for i in range(n):
        tmp *= i+1
    return tmp
````

````python
factorial(8)
````
